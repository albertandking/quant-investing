## 第一段代码：加载行情、算收益、画净值
用内置数据，无需联网。

```python
import matplotlib.pyplot as plt
from qfin import load_close, set_chinese_font

set_chinese_font()
close = load_close()
one = close['STK001']
nav = one / one.iloc[0]
nav.plot(title='STK001 净值曲线（起点归一为 1）')
plt.xlabel('日期'); plt.ylabel('净值'); plt.show()
print(f'区间累计收益：{nav.iloc[-1] - 1:.1%}')
```

### 习题1.4：对比两只股票

```python
pair = close[['STK001','STK010']]
(pair / pair.iloc[0]).plot(title='两只股票净值对比')
plt.xlabel('日期'); plt.ylabel('净值'); plt.show()
```
